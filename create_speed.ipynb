This file will create the data for speed.csv

```r
library(tidyverse)
library(dplyr)  
```

```r
# import the csv file with the data about the crashes in chch into R
chch_crash_data <- read_csv(file = "chch_crashes.csv")
chch_crash_data 
```

```r
# rename the first column (which is the primary key) to crash_id
colnames(chch_crash_data)[1] <- 'speed_id' # this is a more reliable method as it has a problem that the first column is called '...1'
chch_crash_data
```

```r
# create the speed data frame
speed <- select(chch_crash_data , speed_id, advisorySpeed, speedLimit, temporarySpeedLimit)
speed %>% glimpse()
```

```r
# returns a vector containig the column names that contain the word speed  
speed_cols <- function(df){
    
    matches = vector()
    
    # loops through the data frames columns and stores the ones containing speed in the 
    for (i in colnames(df)){
        
        if (str_detect(i, regex("speed", ignore_case = TRUE))){
            matches <- append(matches, i)
            
        }
    }
    speed_data <- select(df, matches)
    return(speed_data)
    
}
speed = chch_crash_data %>% speed_cols() 
speed %>% glimpse()
```

```r
# write the data frame to a csv file 
write.csv(speed, "speed.csv")
```
