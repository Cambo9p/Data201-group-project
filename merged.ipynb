This file will import the crash analysis data and sort it by location
(into the christchurch/canterbury reigon)

```r
library(tidyverse)
library(dplyr)
```

```r
# import the csv data from the file into R 
crash_data <- read_csv(file = "crash_data.csv") 
crash_data
```

```r
unique(crash_data$region)
```

```r
# this function will return a vector containing the matches to the regex 
# which searches through the rows in the dataframe
region_vector <- function(df) {
    
    search_regex = "canterbury|christchurch|chch" # the regex to match the regions - can be changed to add other regions 
    df <- df %>% mutate_at('region', ~replace_na(.,"")) # replace NA values w empty string
    matches = vector()
    for (i in unique(df$region)) {
        
        if (str_detect(i, regex(search_regex, ignore_case = TRUE))){
            matches <- append(matches, i)
            
        }
    }
    matches
}
re_matches <- region_vector(crash_data)
re_matches
```

```r
# now filter the crash data so it only shows crashes in christchurch/canterbury
chch_crashes = crash_data %>% filter(region %in% re_matches)
chch_crashes 
```

now we want to download the chch crash data to create our data model 

```r
# this writes the raw data to the chch_crashes.csv file 
write.csv(chch_crashes, "chch_crashes.csv")
```

```r
# import the csv file with the data about the crashes in chch into R
chch_crash_data <- read_csv(file = "chch_crashes.csv")
chch_crash_data 
```

```r
# rename the first column (which is the primary key) to crash_id
colnames(chch_crash_data)[1] = 'vehicles_id' # this is a more reliable method as it has a problem that the first column is called '...1'
chch_crash_data
```

```r
# crate a df called vehilce that contains all the information about the vehicles involved 
vehicles = select(chch_crash_data, vehicles_id,
                 bicycle, bus, carStationWagon,
                 moped, motorcycle, parkedVehicle,
                 schoolBus, suv, taxi, truck,
                 vehicle, vanOrUtility, strayAnimal)
vehicles %>% glimpse()               
```

```r
# replaces all the NA values with 0 as these are equivalent 
vehicles[is.na(vehicles)] <- 0
vehicles %>% glimpse()
```

```r
# create a column called count that has a count of the vehicles or animals involved 
sliced_vehicles = vehicles[-(1:1)]
count  <-  rowSums(sliced_vehicles)
vehicles  <-  cbind(vehicles, count)
vehicles %>% glimpse()
# be careful with this cell as if you run it more than once it will create another count column with the incorrect values 
```

```r
# write the data frame to a csv file 
write.csv(vehicles, "vehicles.csv")
```

This file will create the data for speed.csv

```r
library(tidyverse)
library(dplyr)  
```

```r
# import the csv file with the data about the crashes in chch into R
chch_crash_data <- read_csv(file = "chch_crashes.csv")
chch_crash_data 
```

```r
# rename the first column (which is the primary key) to crash_id
colnames(chch_crash_data)[1] <- 'speed_id' # this is a more reliable method as it has a problem that the first column is called '...1'
chch_crash_data
```

```r
# create the speed data frame
speed <- select(chch_crash_data , speed_id, advisorySpeed, speedLimit, temporarySpeedLimit)
speed %>% glimpse()
```

```r
# returns a vector containig the column names that contain the word speed  
speed_cols <- function(df){
    
    matches = vector()
    
    # loops through the data frames columns and stores the ones containing speed in the 
    for (i in colnames(df)){
        
        if (str_detect(i, regex("speed", ignore_case = TRUE))){
            matches <- append(matches, i)
            
        }
    }
    speed_data <- select(df, matches)
    return(speed_data)
    
}
speed = chch_crash_data %>% speed_cols() 
speed %>% glimpse()
```

```r
# write the data frame to a csv file 
write.csv(speed, "speed.csv")
```
